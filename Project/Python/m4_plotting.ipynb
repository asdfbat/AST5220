```python
import numpy as np
import matplotlib.pyplot as plt
from PyConstants import Constants

plt.rcParams.update({'font.size':16})
```

```python
data = np.loadtxt("../data/cells.txt")
ells, Cell = data.T

data = np.loadtxt("../data/P.txt")
k, P = data.T

data = np.loadtxt("../data/Theta_l=2.txt")
k, Theta2 = data.T
data = np.loadtxt("../data/Theta_l=20.txt")
k, Theta20 = data.T
data = np.loadtxt("../data/Theta_l=120.txt")
k, Theta120 = data.T
data = np.loadtxt("../data/Theta_l=400.txt")
k, Theta400 = data.T
```

```python
h = 0.7
Mpc = Constants.Mpc
k_Mpc = k*Mpc
```

```python
plt.figure(figsize=(16,10))
plt.loglog(k_Mpc, np.abs(Theta2), label="l=2")
plt.loglog(k_Mpc, np.abs(Theta20), label="l=20")
plt.loglog(k_Mpc, np.abs(Theta120), label="l=120")
#plt.loglog(k, np.abs(Theta400))
plt.legend()
plt.xlabel("k [1/Mpc]")
plt.ylabel("$\Theta_\ell(k)$");
```

```python
plt.figure(figsize=(16,10))
plt.loglog(k_Mpc, Theta2**2/k, label="l=2")
plt.loglog(k_Mpc, Theta20**2/k, label="l=20")
plt.loglog(k_Mpc, Theta120**2/k, label="l=120")
#plt.loglog(k, np.abs(Theta400))
plt.legend()
plt.xlabel("k [1/Mpc]")
plt.ylabel("$\Theta_\ell(k)$");
```
