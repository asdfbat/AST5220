```python
import numpy as np
import matplotlib.pyplot as plt
from PyConstants import Constants

plt.rcParams.update({'font.size':18})
```

```python
h = 0.7
Mpc = Constants.Mpc
```

```python
data = np.loadtxt("../data/cells.txt")
ells, Cell = data.T
```

```python
max_Cell_idx = np.argmax(Cell)
print(ells[max_Cell_idx])

plt.figure(figsize=(10,6))
plt.loglog(ells, Cell)
plt.scatter([ells[max_Cell_idx]], [Cell[max_Cell_idx]], s=100, c="y")
plt.axvline(x=[ells[max_Cell_idx]], c="y", ls="--")
plt.xlabel("$\ell$")
plt.ylabel("$C_\ell\ - \ [\mu K^2]$")
plt.title("$C_\ell \cdot \ell(\ell+1)/2\pi \cdot (10^6 T_{CMB,0})^2$");
plt.yticks([1e2, 1e3, 1e4])
plt.tight_layout()
plt.savefig("../m4_figs/Cell.png", bbox_inches="tight")
```

```python
data_SW = np.loadtxt("../data/cells_SW.txt")
ells, Cell_SW = data_SW.T

data_ISW = np.loadtxt("../data/cells_ISW.txt")
ells, Cell_ISW = data_ISW.T

data_Dop = np.loadtxt("../data/cells_Dop.txt")
ells, Cell_Dop = data_Dop.T

data_Quad = np.loadtxt("../data/cells_Quad.txt")
ells, Cell_Quad = data_Quad.T
```

```python
plt.figure(figsize=(10,6))
plt.loglog(ells, Cell, label="SUM")
plt.loglog(ells, Cell_SW, label="SW")
plt.loglog(ells, Cell_ISW, label="ISW")
plt.loglog(ells, Cell_Dop, label="Dop")
plt.loglog(ells, Cell_Quad, label="Quad")

plt.xlabel("$\ell$")
plt.title("Components of $C_\ell \cdot \ell(\ell+1)/2\pi \cdot (10^6 T_{CMB,0})^2$");
plt.ylabel("$C_\ell \ - \ [\mu K^2]$")
#plt.yticks([1e2, 1e3, 1e4])
plt.ylim(1e-2, 1e4)
plt.legend();
plt.tight_layout()
plt.savefig("../m4_figs/Cell_all.png", bbox_inches="tight");
```

```python
data = np.loadtxt("../data/P.txt")
h = 0.7
k, P = data.T
P *= h
k_max_idx = np.argmax(P)
```

```python
k_Mpc = k*Mpc
h_k_Mpc = k_Mpc/h

x_ax = h_k_Mpc
y_ax = P*(h/Mpc)**3

plt.figure(figsize=(10,6))
plt.loglog(x_ax, y_ax)
plt.axvline(x=h_k_Mpc[k_max_idx], ls="--", c="y");
plt.scatter([h_k_Mpc[k_max_idx]], [y_ax[k_max_idx]], c="y", s=100)
plt.xlabel("k - [h/Mpc]")
plt.title("P - $[Mpc^3/h^3]$")
plt.tight_layout()
plt.savefig("../m4_figs/MatterPower.png", bbox_inches="tight")

k_Mpc[k_max_idx], h_k_Mpc[k_max_idx]
```
